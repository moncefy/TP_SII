```python
import random
import math
import numpy as np
```

```python
#F1 
def F1(x):
    resulta = 0.0
    for i in range(len(x)):
        resulta = resulta + x[i] * x[i]
    return resulta
```

```python
def random_solution(D, min_val, max_val):
    x = []
    for _ in range(D):
        x.append(random.uniform(min_val, max_val))
    return x
```

```python
# ntestiw ntestiw
D = 30
x_random = random_solution(D, -100, 100)
print("F1(x_random) =",x_random) 
```
