```python
# Liste donnée par le prof
x = [-27.81, -71.96, -47.13, 54.63, -86.58, -96.77, 63.39, 75.60,
     -39.94, -45.13, 90.77, 70.68, 36.61, 18.50, -46.11, -91.04,
     -34.74, 94.34, 61.98, -77.94, -78.75, -3.11, -2.81, 80.69,
     -76.95, 43.46, 3.65, -26.73, 49.26, 0.72]

# F2 = somme(|xi|) + produit(|xi|)
def F2(x):
    somme = 0
    produit = 1
    
    for xi in x:
        somme += abs(xi)
        produit *= abs(xi)
    
    return somme + produit

resultat = F2(x)
print("F2 =", resultat)
```
