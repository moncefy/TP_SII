```python
# Evaluate all lines of TP2/Population_F1-UM.csv with F1 from TP1 (Sphere): sum(x_i^2)
df_f1 = pd.read_csv("TP2/Population_F1-UM.csv", sep=';', header=None)
scores_f1 = (df_f1 ** 2).sum(axis=1)

best_idx = int(scores_f1.idxmin())
worst_idx = int(scores_f1.idxmax())

print("Rows:", len(df_f1), "Dim:", df_f1.shape[1])
print("Best line (1-based):", best_idx + 1, "F1:", float(scores_f1.iloc[best_idx]))
print("Worst line (1-based):", worst_idx + 1, "F1:", float(scores_f1.iloc[worst_idx]))
print("Best vector:", df_f1.iloc[best_idx].tolist())
print("Worst vector:", df_f1.iloc[worst_idx].tolist())
```
