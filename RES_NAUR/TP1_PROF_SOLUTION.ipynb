## 1. IMPORT DES BIBLIOTHÈQUES

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. DATASET

```python
# Toy Dataset : y = 2x + 1 + bruit
np.random.seed(42)

X = np.linspace(0, 10, 50)
bruit = np.random.randn(50) * 1.5
y = 2 * X + 1 + bruit
```

## 3. VISUALISATION DES DONNÉES

```python
plt.scatter(X, y)
plt.title("Dataset initial")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## 4. INITIALISATION DES PARAMÈTRES

```python
theta_0 = 0.0   # biais
theta_1 = 0.0   # pente

learning_rate = 0.01
epochs = 10
```

## 5. FONCTIONS DE LA DESCENTE DU GRADIENT

```python
def prediction(x, theta_1, theta_0):
    return theta_1 * x + theta_0


def cost_function(X, y, theta_1, theta_0):
    n = len(X)
    y_pred = prediction(X, theta_1, theta_0)
    return (1/n) * np.sum((y_pred - y)**2)


def gradients(X, y, theta_1, theta_0):
    n = len(X)
    y_pred = prediction(X, theta_1, theta_0)

    d_theta_1 = (2/n) * np.sum((y_pred - y) * X)
    d_theta_0 = (2/n) * np.sum(y_pred - y)

    return d_theta_1, d_theta_0
```

```python
# Desente du gradient
cost_history = []
cost = cost_function(X, y, theta_1, theta_0)

for epoch in range(epochs):

    if epoch % 2 == 0:
        print(f"Epoch {epoch} | Cost = {cost:.4f} | theta_1 = {theta_1:.4f} | theta_0 = {theta_0:.4f}")
        # la droite apprise
        y_pred = prediction(X, theta_1, theta_0)

        plt.scatter(X, y, label="Données réelles")
        plt.plot(X, y_pred, color="red", label="Droite apprise")
        plt.legend()
        plt.title("Régression linéaire apprise par descente de gradient")
        plt.show()
        
    d_theta_1, d_theta_0 = gradients(X, y, theta_1, theta_0)

    # mise à jour des paramètres
    theta_1 = theta_1 - learning_rate * d_theta_1
    theta_0 = theta_0 - learning_rate * d_theta_0

    # calcul coût
    cost = cost_function(X, y, theta_1, theta_0)
    cost_history.append(cost)

```

```python
# Affichage des paramètres finaux
print("\nParamètres appris :")
print("theta_1 =", theta_1)
print("theta_0 =", theta_0)
```

## 6. VISUALISATION du cout

```python
# la courbe de convergence
plt.plot(cost_history)
plt.title("Évolution de la fonction de coût")
plt.xlabel("Epoch")
plt.ylabel("Coût")
plt.show()
```
